```python
import pandas as pd
import matplotlib.pyplot as plt

# load the CSV file
df = pd.read_csv("../output/species_counts.csv")

# define bins and labels
bins = [0, 200, 400, 600, 800, 1000]
labels = ['1–200', '201–400', '401–600', '601–800', '801–1000']

# categorize counts into bins
df['count_bin'] = pd.cut(df['count'], bins=bins, labels=labels, right=True)

# count how many species fall into each bin
bin_counts = df['count_bin'].value_counts().sort_index()

# plot
plt.figure(figsize=(8, 6))
bin_counts.plot(kind='bar', color='lightgreen', edgecolor='black')
plt.ylabel("Number of Species")
plt.xlabel("Photo Count Range")
plt.title("Distribution of Species by Photo Count")
plt.tight_layout()
plt.show()
```
